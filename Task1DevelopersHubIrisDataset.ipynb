# Task 1: Exploring and Visualizing a Simple Dataset

# Problem Statement:
The Iris dataset is one of the most well-known datasets in data science and machine learning. It contains measurements of different iris flower species, enabling the study of relationships between flower features and species classification. This task involves loading, inspecting, and visualizing the dataset to understand data distributions, correlations between features, and potential outliers, providing a foundation for further analysis.

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets

# metadata
print(iris.metadata)

# variable information
print(iris.variables)
```

```python
# Basic dataset inspection

# Check shapes of features and targets
print("Shape of X (features):", X.shape)
print("Shape of y (target):", y.shape)

# First 5 rows of features
print("\nFirst 5 rows of features:\n", X.head())

# First 5 rows of target values
print("\nFirst 5 rows of targets:\n", y.head())

# Column names
print("\nFeature column names:", X.columns.tolist())
print("Target column name:", y.columns.tolist())

# Dataset info
print("\nFeatures Info:")
print(X.info())

# Statistical summary of numerical features
print("\nStatistical Summary:\n", X.describe())
```

```python
# Data visualization and exploration
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Combine features and target for easier plotting
df = pd.concat([X, y], axis=1)

# Scatter plot - Sepal Length vs Sepal Width colored by species
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x="sepal length", y="sepal width", hue="class", palette="Set1")
plt.title("Sepal Length vs Sepal Width by Species")
plt.show()
```

```python
# Histograms for each numerical feature
df.hist(figsize=(8, 6), bins=15, edgecolor="black")
plt.suptitle("Feature Distributions", y=1.02)
plt.show()
```

```python
# Box plots to check outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=df.drop(columns="class"))
plt.title("Box Plot of Iris Features")
plt.show()
```

# Scatter Plot Insights:
The scatter plot clearly shows that species are well-separated in terms of petal measurements, while sepal measurements have some overlap between species. Petal length and width in particular form distinct clusters, indicating they are highly informative for classification.

# Histogram Insights:
The histograms reveal that petal length and petal width have multi-modal distributions, reflecting the differences between species groups. In contrast, sepal length is more evenly distributed, and sepal width is slightly skewed with a peak around 3 cm. These patterns confirm that petal features vary more across species.

# Box Plot Insights:
The box plots show petal features have the largest spread, with minimal overlap between species, while sepal measurements are more compact. Sepal width contains a few outliers, but the rest of the dataset appears clean. This suggests strong predictive potential from petal measurements and relatively consistent sepal measurements.
